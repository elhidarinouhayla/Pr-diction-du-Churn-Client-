### Visualisation :

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
import seaborn as sns 
```

```python
file['TotalCharges'] = pd.to_numeric(file["TotalCharges"], errors= 'coerce')
file.isnull().sum()
mean = file['TotalCharges'].mean()
file['TotalCharges'] = file['TotalCharges'].fillna(mean)
colonneNum = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
file['Churn'] = le.fit_transform(file['Churn'])
correlation = file[colonneNum].corr()
correlation
sns.heatmap(correlation, annot= True, cmap= 'coolwarm')

```

```python

# colonnes_ca = file.drop(columns=colonneNum)
# colonnes_ca
# for colonne in colonnes_ca:
#    sns.countplot(data=file, x=colonne, hue='Churn')

colonneNum = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
colonnes_ca = file.drop(columns=colonneNum).columns   # ✅ Get only column names

for colonne in colonnes_ca:
    plt.figure(figsize=(6,4))
    sns.countplot(data=file, x=colonne, hue='Churn')
    plt.title(f'Distribution of {colonne} by Churn')
    plt.xticks(rotation=30)
    plt.show()
```
