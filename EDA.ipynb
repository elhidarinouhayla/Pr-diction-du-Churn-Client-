### Visualisation :

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
import seaborn as sns 
```

```python
colonneNum = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'MonthlyCharges', 'Churn']
correlation = file[colonneNum].corr()['Churn']
corr = correlation['Churn']
corr
```

```python
# 1-Nombre de clients qui ont quitté (Yes) ou resté (No)

le = LabelEncoder()
for colonne in file.columns:
    if file[colonne].dtype == 'object':
        file[colonne] = le.fit_transform(file[colonne])
corr = file.corr()
# corr
plt.figure(figsize=(14,14))
sns.heatmap(corr, annot= True, cmap= 'coolwarm')
    


    


```
