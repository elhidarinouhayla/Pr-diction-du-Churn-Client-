### Visualisation :

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
import seaborn as sns 
```

```python
#  1-correlation des variables numeriques :

file['TotalCharges'] = pd.to_numeric(file["TotalCharges"], errors= 'coerce')
file.isnull().sum()
mean = file['TotalCharges'].mean()
file['TotalCharges'] = file['TotalCharges'].fillna(mean)
colonneNum = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']
file['Churn'] = le.fit_transform(file['Churn'])
correlation = file[colonneNum].corr()
correlation
sns.heatmap(correlation, annot= True, cmap= 'coolwarm')
plt.title("Matrice de correlation")
plt.show()

```

```python
# 2-création de countplots pour les variables catégorielles par rapport à Churn"

colonnes_ca = file.drop(columns=colonneNum + ['customerID']).columns   
for colonne in colonnes_ca:
    plt.figure(figsize=(8,6))
    sns.countplot(data=file, x=colonne, hue='Churn')
    plt.title(f'Distribution of {colonne} by Churn')
    plt.show()
```
